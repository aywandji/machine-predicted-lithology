## Data Analysis/vizualisation

```python
train_df = pd.read_csv(data_folder + "train.csv",sep=";")
test_df = pd.read_csv(data_folder + "test.csv",sep=";")
```

```python
train_df.describe()
```

```python
train_df.info()
```

```python
### get wells drilling zone geometry
wells_positions_df = train_df.groupby("WELL").agg({"X_LOC":["min","max","mean"],"Y_LOC":["min","max","mean"]})
wells_positions_df.columns = ["_".join(x) for x in wells_positions_df.columns.ravel()]
wells_positions_df["well_width"] = wells_positions_df.X_LOC_max - wells_positions_df.X_LOC_min
wells_positions_df["well_height"] = wells_positions_df.Y_LOC_max - wells_positions_df.Y_LOC_min
wells_positions_df["well_max_surface"] = wells_positions_df.well_height * wells_positions_df.well_width

test_wells_positions_df = test_df.groupby("WELL").agg({"X_LOC":["min","max","mean"],"Y_LOC":["min","max","mean"]})
test_wells_positions_df.columns = ["_".join(x) for x in test_wells_positions_df.columns.ravel()]
test_wells_positions_df["well_width"] = test_wells_positions_df.X_LOC_max - test_wells_positions_df.X_LOC_min
test_wells_positions_df["well_height"] = test_wells_positions_df.Y_LOC_max - test_wells_positions_df.Y_LOC_min
test_wells_positions_df["well_max_surface"] = test_wells_positions_df.well_height * test_wells_positions_df.well_width

wells_positions_df
```

```python
fig,axes = plt.subplots(1,2,figsize=(15,5))
axes = axes.flatten()

sns.scatterplot(x="X_LOC_mean",y="Y_LOC_mean",marker="o",size="well_max_surface",data=wells_positions_df,ax = axes[0])
sns.scatterplot(x="X_LOC_mean",y="Y_LOC_mean",marker="o",size="well_max_surface",color="red",data=test_wells_positions_df,ax = axes[0])

sns.scatterplot(x="well_width",y="well_height",marker="x",data=wells_positions_df,ax = axes[1])




```

### *COMMENTS*
- All x,y samples from the same well fall close to each other (fair enough)
- train and test data don't have wells in common so our model shouldn't learn from specific well characteristics but should generalize to data coming from any well position (this is also fair enough)
- there are 6 wells with surface ==0 (meaning that the drilling was fully vertical?). 

```python
def plot_well_logs(well_name,measures_list,train_df,figsize=(25,8)):
    """
    Plot well logs specified in measures_list
    """
    well_measures_df  = train_df.loc[train_df.WELL==well_name]
    fig,axes = plt.subplots(1,len(measures_list)+1,figsize=figsize)
    axes = axes.flatten()
    for i,col in enumerate(measures_list):
        axes[i].plot(well_measures_df[col],well_measures_df.DEPTH_MD)
        axes[i].set_title(col)
        axes[i].invert_yaxis()

    axes[-1].imshow(well_measures_df.FORCE_2020_LITHOFACIES_LITHOLOGY.values.reshape((len(well_measures_df),1)),aspect="auto",cmap="tab20")
    axes[-1].set_title("lithofacies")
    plt.show()
```

```python
#plot some logs(all features) per well to vizualise
well_name = train_df.WELL.unique()[8]
measures_list = train_df.columns.values[7:-2]

plot_well_logs(well_name,measures_list,train_df,figsize=(25,8))
```

```python
#check features distribution per well to define how to normalize data
def percentile(n):
    def percentile_(x):
        return np.percentile(x, n)
    percentile_.__name__ = 'percentile_%s' % n
    return percentile_

min_max_df = train_df[list(measures_list) + ["WELL"]].groupby("WELL").agg(["min","mean","median",percentile(75),percentile(95),"max"])
min_max_df.columns = ["_".join(x) for x in min_max_df.columns.values.ravel()]
min_max_df
```

```python
nb_cols = 5
nb_rows = int(np.ceil(len(measures_list) / nb_cols))
fig,axes = plt.subplots(nb_rows,nb_cols,figsize=(25,15))
axes = axes.flatten()
for i,col in enumerate(measures_list):
    sns.scatterplot(x=col+"_min",y=col+"_percentile_75",data=min_max_df,ax=axes[i])
```

### *COMMENTS*
- There are often some spike values in features series. Maybe they represent rock layer transition? or they are juste errors(outliers)?
- Thus, we can try to mitigate the potential errors by setting the max value of a feature to 1.5*IQR from the 75th percentile (after removing Nan values)
- ALL features have similar range whatever the well we consider, so we can normalize those features across all wells.

## Check null values distribution

```python
nan_values_df = train_df[measures_list].isna().mean().sort_values()
```

```python
plt.figure(figsize=(20,5))
splot = sns.barplot(x=nan_values_df.index,y=nan_values_df)
splot.set_title("percentage of null values per measurement (all wells data)")
plt.show()
```

```python
nan_values_per_well_df = train_df.isna()[measures_list]
nan_values_per_well_df["WELL"] = train_df.WELL
nan_values_per_well_df = nan_values_per_well_df.groupby("WELL").mean()
```

```python
#distribution over the wells of null values percentage per feature
nb_cols = 10
nb_rows = int(np.ceil(len(measures_list) / nb_cols))
fig,axes = plt.subplots(nb_rows,nb_cols,figsize=(30,10))
axes = axes.flatten()
for i,col in enumerate(measures_list):
    sp = sns.boxplot(y=col,data=nan_values_per_well_df,ax=axes[i])
plt.suptitle("wells percentage of null values per feature")
plt.show()
```

```python
##order wells per missing values count overall
missing_data_threshold = 0.6
wells_missing_data = nan_values_per_well_df.mean(axis=1).sort_values()
wells_missing_data.plot(kind="bar",figsize=(30,5))
plt.axhline(y=missing_data_threshold, color='green', linestyle='--',label="missing percentage threshold")
plt.axhline(y=wells_missing_data.mean(), color='blue', linestyle='-',label="mean of missing data percentage per well")
plt.title("percentage of missing data per well")
plt.legend()
plt.show()
```

```python
wells_missing_data[wells_missing_data>0.6]
```

```python
##number of features per well with N% of missing values
feature_missing_data_threshold = 0.75

wells_missing_data2 = (nan_values_per_well_df>feature_missing_data_threshold).sum(axis=1).sort_values()
wells_missing_data2.plot(kind="bar",figsize=(30,5))
plt.axhline(y=nan_values_per_well_df.shape[1]//2, color='green', linestyle='--',label="half number of measurements")
plt.axhline(y=wells_missing_data2.mean(), color='blue', linestyle='-',label="average value per well")
plt.title("number of measurements with more than {}% of missing data per well".format(feature_missing_data_threshold*100))

plt.legend()
plt.show()
```

```python
#Check null values distribution per z, depth_md (depth) to verify 
#if there is a pattern that could be due to depth
measures_list = train_df.columns.values[7:-2]

nb_intervals = 20
cat_z_df = train_df.loc[:,list(measures_list)+["Z_LOC"]]
cat_z_df["z_inter"] = pd.cut(cat_z_df.Z_LOC,nb_intervals, labels=range(nb_intervals,0,-1))

null_per_z_df = cat_z_df.loc[:,list(measures_list)].isna()
null_per_z_df["z_inter"] = cat_z_df["z_inter"]
null_per_z_df = null_per_z_df.groupby("z_inter").mean()
```

```python
plt.figure()
nb_rows = 4
ncols = int(np.ceil(len(measures_list) /nb_rows))
fig,axes = plt.subplots(nb_rows,ncols,figsize=(25,15))
axes = axes.flatten()
for i,col in enumerate(measures_list):
    axes[i].plot(null_per_z_df[col].values,list(null_per_z_df.index.values))
    axes[i].set_title(col)
    axes[i].axvline(x=0.6,color="green")
    axes[i].invert_yaxis()
fig.suptitle("Percentage of null values according to Z_COL",fontsize=50)
plt.show()
```

```python
#percentage of null values per lithofacy
null_values_per_label = train_df.loc[:,list(measures_list)].isna()
null_values_per_label["FORCE_2020_LITHOFACIES_LITHOLOGY"] = train_df["FORCE_2020_LITHOFACIES_LITHOLOGY"]
null_values_per_label["FORCE_2020_LITHOFACIES_CONFIDENCE"] = train_df["FORCE_2020_LITHOFACIES_CONFIDENCE"]
null_values_per_label = null_values_per_label.groupby("FORCE_2020_LITHOFACIES_LITHOLOGY").mean().mean(axis=1)
```

```python
null_values_per_label
```

```python
null_values_per_label.plot(kind="bar",figsize=(20,5),title="Null values percentage per lithofacies")
```

### *COMMENTS*
- The percentage of measurements null values according to z show that some very sparse measurements like SGR,DTS,RMIC,ROPA and many more become less sparse around -3545 and -4074 of Z_LOC.

- A lot of measurements (NHPI, PEF,BS) are very sparse when Z_LOC is very high (-111 to -1432). Is it due to the fact that they rae not useful to classify lithofacies when closer to ground surface? 
- Same behaviour when closer to the bottom (PEF,ROP)

- Maybe we should train a specific model for near ground surface/bottom surface lithofacies, for Z_LOC around -3545 and -4074 and a model for the remaining Z_LOC.

- percentage of null values per lithofacy class is similar and close to 0.4. Exceptions are "99000" and "88000" lithofacies with 60% of null values. So the sparsity is a priori not related to lithofacies.

### Features correlation

```python
# Compute the correlation matrix
features_correlation = train_df[numeric_features_col].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(features_correlation, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(20, 13))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sp = sns.heatmap(features_correlation, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
sp.set_title("correlation matrix")
```

### *COMMENTS*
- DEPTH_MD and Z_LOC are almost identical. We will keep only one of them.
- Wells data are acquired in very different depth intervals. min,max values of DEPTH_mD are different for each well. Also the number of samples per well is very different. We should consider this during training.

- RSHA, RMED,RDEP bring out very clearly lithofacies 86000, 88000 (65% of null measurements) and 93000. 

- Same thing for BS and 99000 (60% of null measurements)
- Same thing for DTS,RXO and 86000 (60% of null measurements)

- they all have some outliers values (Q3,1 +- 1.5IQR). We can either ignore this or threshold those outliers. This is because those outliers can be only numeric spikes (errors).

- lithofacies 70032 (44% of null features),74000(47% of null features) and 88000(65% of null features) have the lowest labels confidence values. Therefore maybe we shouldn't try to mimic experts classification for those labels (by giving a low weight to the samples of interest)

- From features correlation matrix we saw that many features are highly correlated (RHOB, NPHI, DTC, DTS,...). This can be used to fill null values.
